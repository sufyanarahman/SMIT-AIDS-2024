```python
data = pd.read_csv("diabetes.csv")
```

# Step2: Separate Input and out features

```python
X = data.drop('Outcome',axis=1)
y = data['Outcome']
```

# Step3: Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0, test_size=.20)
```

# Step4: Model Selection
       - Classification problem in hand

# Common Classification Algorithms:
# Logistic Regression:
                A statistical method for predicting the probability of an event occurring. 
# Support Vector Machines (SVM):
                A powerful algorithm that finds the optimal hyperplane to separate data points into different classes. 
# Decision Trees:
                Algorithms that use a tree-like structure to make predictions based on a series of decisions. 
# Random Forest:
                An ensemble method that combines multiple decision trees to improve accuracy. 
# K-Nearest Neighbors (KNN):
                An algorithm that classifies data points based on the majority class of their nearest neighbors. 
# Neural Networks:
                Complex models inspired by the structure of the human brain, capable of learning complex patterns. 

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression()
```

# Step5: Model Fit

```python
logr.fit(X_train,y_train)
```

```python
y_pred_train = logr.predict(X_train)
y_train
y_pred_train
```
