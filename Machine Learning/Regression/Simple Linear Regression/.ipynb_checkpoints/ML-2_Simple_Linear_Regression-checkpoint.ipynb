# Step:1 Data Load

```python
import pandas as pd
```

```python
data = pd.read_csv("weight-height.csv")
```

```python
data.shape
```

```python
data.Gender = [1 if g=="Male" else 0 for g in data.Gender]
data
```

```python
data.head()
```

# Step:2 Separate Input Data and Output Data
    - Data and Labels

```python
X = data[["Height"]]
y = data[["Weight"]]
```

# Step:3 Split Data Into Training Data and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0, test_size= .40)
```

```python
X_train
```

# Step:4 Model Selection

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

# Step:5 Model Training (Fit karna)

```python
# Supervised Learning data + labels
model.fit(X_train,y_train)
```

# Step:6 Prediction On Training Data

```python
y_predict_training = model.predict(X_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
pd.DataFrame({"y_train":y_train.values.reshape(6000,),"y_pred":y_predict_training.reshape(6000,)})
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2_score(y_train,y_predict_training)
```

```python
mean_absolute_error(y_train,y_predict_training)
```

```python
mean_squared_error(y_train,y_predict_training)
```

# Step:7 Prediction On Test Data

```python
y_pred_test = model.predict(X_test)
```

```python
r2_score(y_test,y_pred_test)
```

```python
mean_absolute_error(y_test,y_pred_test)
```

```python
def predict_weight(height):
    return model.predict(height) * .453592
```
